## Data Cleaning

```python
import pandas as pd
import numpy as np
df = pd.read_csv("./Car_Insurance_Claim.csv")
df.head(5)
```

Removing first column (It has the same values like index column)

```python
df = df.drop(df.columns[0], axis=1)
df.head(5)
```

We can also notice that there are a lot of "none" values. In some columns it can be interpreted as "Not set" or NaN, but it can also mean 0 in some cases or, for example, in "EDUCATION" column it can mean "No education".

```python
# In these columns "none" values can be interpreted as NaN or "Not set"
columns_to_nan = ["ID", "AGE", "GENDER", "RACE", "INCOME", "CREDIT_SCORE", "VEHICLE_YEAR", "POSTAL_CODE", "VEHICLE_TYPE"]
# In these columns "none" values can be interpreted as 0
columns_to_zero = ["VEHICLE_OWNERSHIP", "MARRIED", "CHILDREN", "ANNUAL_MILEAGE", "SPEEDING_VIOLATIONS", "DUIS", "PAST_ACCIDENTS", "OUTCOME"]

df[columns_to_nan] = df[columns_to_nan].replace("none", np.NaN)
df[columns_to_zero] = df[columns_to_zero].replace("none", 0)

# Special cases
df["DRIVING_EXPERIENCE"] = df["DRIVING_EXPERIENCE"].replace("none", "0y")
df["EDUCATION"] = df["EDUCATION"].replace("none", "No education")
```

Finally, all the values of given dataset are represented as strings. Some columns contain numeric data, so let's transform them into numeric data type.

```python
print("Before transformation: ", df.loc[0, "CREDIT_SCORE"] + df.loc[0, "CREDIT_SCORE"])

numeric_columns = ["CREDIT_SCORE", "VEHICLE_OWNERSHIP", "MARRIED", "CHILDREN", "ANNUAL_MILEAGE", "SPEEDING_VIOLATIONS", "DUIS", "PAST_ACCIDENTS", "OUTCOME"]

for col in numeric_columns:
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("After transformation: ", df.loc[0, "CREDIT_SCORE"] + df.loc[0, "CREDIT_SCORE"])
```

## Data Visualization

### Visualizing overall data

Displaying numeric data

```python
df.hist(figsize=(30, 30), bins=20, color="#107009AA")
```

Displaying non-numeric data

```python
import matplotlib.pyplot as plt

non_numeric_columns = ["AGE", "GENDER", "RACE", "DRIVING_EXPERIENCE", "EDUCATION", "INCOME", "VEHICLE_YEAR", "VEHICLE_TYPE"]
colors = ["orange", "green", "red", "blue"]
```

```python
fig, axes = plt.subplots(4, 2, figsize=(30, 40))
for i, col in enumerate(non_numeric_columns):
    values = df[col].value_counts()
    n_row = i // 2
    n_col = i % 2

    values.plot(kind="bar", ax=axes[n_row, n_col], color=colors[i % 4])
    axes[n_row, n_col].set_title(col)
    axes[n_row, n_col].set_xlabel(None)
```

### Visualizing data with positive outcomes

Displaying numeric data

```python
df[df["OUTCOME"] == 1].hist(figsize=(30, 30), bins=20, color="#107009AA")
```

Displaying non-numeric data

```python
fig, axes = plt.subplots(4, 2, figsize=(30, 40))
for i, col in enumerate(non_numeric_columns):
    values = df[df["OUTCOME"] == 1][col].value_counts()
    n_row = i // 2
    n_col = i % 2

    values.plot(kind="bar", ax=axes[n_row, n_col], color=colors[i % 4])
    axes[n_row, n_col].set_title(col)
    axes[n_row, n_col].set_xlabel(None)
```

## Conclusions

Making conclusions about the dataset

- CREDIT_SCORE: Most of the positive outcomes have 0.4 credit score, although the most frequent in dataset is 0.6
- AGE: The most frequent age range in dataset is 26-39, while in positive outcomes it is 16-25
- GENDER: In the overall data the number of males and females is almost equal, but in positive outcomes there are more males
- RACE: Both positive outcomes and the overall data have the same distribution - "majority" is the most frequent value
- DRIVING_EXPERIENCE: Both positive outcomes and the overall data have the same distribution - "between 0 and 9 years" is the most frequent value
- EDUCATION: In most of the overall data the education is either "high school" or "university", while in data with positive outcomes it is mostly "high school"
- INCOME: The upper class level income is the most frequent in the dataset, while poverty level income is the most frequent in data with positive outcomes
- VEHICLE_OWNERSHIP: Most of the positive outcomes have no vehicle ownership, although in the overall data most of the values are "yes"
- VEHICLE_YEAR: Both positive outcomes and the overall data have the same distribution - "before 2015" is the most frequent value
- MARRIED: While the overall data has the same distribution of "yes" and "no", the most frequent value in positive outcomes is "no"
- CHILDREN: Both positive outcomes and the overall data have the same distribution - more people that have children
- ANNUAL_MILEAGE: Both positive outcomes and the overall data have the same distribution - "mileage around 11000" is the most frequent value
- VEHICLE_TYPE: Both in the overall data and in data with positive outcomes sedan is much more popular than sports car
- SPEEDING_VIOLATIONS: In most of the overall data the number of speeding violations is between 0 and 2, while in data with positive outcomes it is almost always 0
- DUIS: Both in the overall data and in data with positive outcomes the number of DUIS is mostly 0
- PAST_ACCIDENTS: In most of the overall data the number of past accidents is between 0 and 2, while in data with positive outcomes it is almost always 0
- OUTCOME: Negative outcomes are more frequent than positive ones (~7000 records vs ~3000 records)
