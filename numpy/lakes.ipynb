```python
import numpy as np 
import random
import matplotlib.pyplot as plt
```

```python
np.random.seed(random.randint(100, 2000))
a = np.random.randint(1, 100, 20)
a
```

```python
plt.plot (a)
```

```python
def find_deepest_lake(a):
    max_distance = 0 # used just for you to compare
    max_points = []

    for i in range(len(a)):
        distance = [0,0]
        points = []

        for j in range(i, 0, -1):
            p1 = a[j]
            p2 = a[j-1]
            points.append(j)
            
            if (p2 < p1):
                break

            if (j-1 == 0):
                points.append(j-1)
    
            distance[0] += p2-p1

        points = list(reversed(points))

        for j in range(i, len(a)-1):
            p1 = a[j]
            p2 = a[j+1]
            points.append(j)
            
            if (p2 < p1):
                break

            if (j+1 == len(a)-1):
                points.append(j+1)
    
            distance[1] += p2-p1
        if (min(distance) > max_distance):
            max_distance = min(distance)
            max_points = points
    return max_points
```

```python
res = find_deepest_lake(a)
plt.plot(a)
plt.plot(res, [a[i] for i in res], c='red')
```
